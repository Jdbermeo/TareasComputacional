Lo primero es cargar los datos del encefalograma en la matriz data 400x24 (24 señales y 400 datos)

```python
data = np.loadtxt("sampled+ma0844az_1-1+_data.txt") 
```

```python
from scipy.fftpack import fft, fftfreq, ifftimport matplotlib.pyplot as plt
```

Aquí graficamos todas las señales del encefalograma, la señal 1 y la señal 21 por independiente para poder ver claramente el comportamiento de estas

```python
n, col = shape(data)r = linspace(0,n,n)  figure(figsize=(18,6)) #Esto es para que las gráficas no queden muy angostas y se pueda leer bien    plt.subplot(1,3,1)plt.plot(r, data)         #Graficamos todas las 24 señalesplt.ylabel(u'Amplitud pulsos')plt.xlabel('Tiempo (minutos)')plt.subplot(1,3,2)plt.plot(r, data[:,0])    #Graficamos la señal 1plt.ylabel(u'Amplitud pulsos')plt.xlabel('Tiempo (minutos)')plt.subplot(1,3,3)plt.plot(r, data[:,20])   #Graficamos la señal 21plt.ylabel(u'Amplitud pulsos')plt.xlabel('Tiempo (minutos)')
```

En esta parte, hallaremos la transformada de Fourier para cada señal y las guardamos en fft_amplitudes 400x24, y recuperamos la frecuencia en freq 400x1. Finalmente graficamos el espectro de potencias contra la frecuencia para todas las señales, para la señal 1 y la señal 21  

```python
fft_amplitudes = empty([n,col])for j in range(col):    fft_amplitudes[:,j] = fft(data[:,j])freq = fftfreq(n,1)figure(figsize=(22,6))                plt.subplot(1,3,1)plt.plot(freq, np.abs(fft_amplitudes[:,:])**2)             #gráfica de espectro de potencias para todas las señalesplt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')plt.subplot(1,3,2)plt.plot(freq, np.abs(fft_amplitudes[:,0])**2)        #gráfica de espectros de potencias para la señal 1plt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')savefig('espectro1.jpeg')plt.subplot(1,3,3)  plt.plot(freq, np.abs(fft_amplitudes[:,20])**2)       #gráfica de espectros de potencias para la señal 21plt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')savefig('espectro21.jpeg')
```

El siguiente algoritmo es para coger únicamente los valores positivos de las frecuencias para que de esta manera hacer un mejor análisis de la función. Esto se puede hacer visto que las señales son periódicas y por ende, lo que hay a un lado lo hay en el otro

```python
half_n = np.ceil(n/2.0)fft_amplitudes_half = empty([n/2,col], dtype=complex)freq_half = freq[:half_n]for i in range(24):    fft_amplitudes_half[:,i] = fft_amplitudes[:half_n,i]    figure(figsize=(22,6))    plt.subplot(1,3,1)plt.plot(freq_half, np.abs(fft_amplitudes_half)**2)           #gráfica de espectro de potencias para todas las señalesplt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')plt.subplot(1,3,2)plt.plot(freq_half, np.abs(fft_amplitudes_half[:,0])**2)        #gráfica de espectros de potencias para la señal 1plt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')plt.subplot(1,3,3)  plt.plot(freq_half, np.abs(fft_amplitudes_half[:,20])**2)       #gráfica de espectros de potencias para la señal 21plt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')
```

Aquí nos quedamos con los 10 valores más grandes de los espectros de potencia de cada señal. Para esto, recurrimos a un método de numpy que se llama .argsort() el cual nos guarda en x los índices de los valores más pequeños. Con el (-fft_amlp) buscamos que .argsort() nos busque los máximos. Finalmente, los fft_amp[x[j,i],i] para j > 19, los hacemos 0. Elijo 19 (20 valores) y no 9 (10 valores) porque cada valor máximo se repite debido a la frecuencia negativa al ser una función periódica. 

```python
fft_amp = np.abs(fft_amplitudes)x = empty((n,col))for i in range(24):                       #loop para determinar los índices en los que se encuentran los valores máximos    x[:,i] = (-fft_amp[:,i]).argsort()for i in range(24):                       #loop para hacer los valores 0.0 a aquellos que no corresponden a los 10 más grandes    for j in range(n):            if j > 19:                fft_amp[x[j,i],i] = 0.0            print fft_amp[:,0]                        #Imprimimos la nueva matriz fft_amp con unicamente los 10 valores más grandes
```

Graficamos los espectros de potencias con los 10 valores más grandes para todos los espectros de todas las señales, la señal 0 y la señal 20

```python
figure(figsize=(22,4))plt.subplot(1,3,1)plt.plot(freq, fft_amp**2)                #gráfica de espectros de potencias para las señalesplt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')plt.subplot(1,3,2)plt.plot(freq, fft_amp[:,0]**2)           #gráfica de espectros de potencias para la señal 1plt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')savefig('maximos1.jpeg')plt.subplot(1,3,3)plt.plot(freq, fft_amp[:,20]**2)          #gráfica de espectros de potencias para la señal 21plt.ylabel(u'Espectro de potencias')plt.xlabel('Frecuencia')savefig('maximos21.jpeg')
```

Ahora hallamos la inversa de la transformada para poder reconstruir la señal. Para esto usamos la función ifft(). Y graficamos para todas las señales, la señal 1 y la señal 21, las señales antiguas (señales de data[ ]) y las señales reconstruidas (inverse[ ]) para poderlas comparar.

```python
inverse = empty((n,col))for i in range(24):        #loop para hallar la inversa para cada transformada de Fourier de cada señal   inverse[:,i] = ifft(fft_amp[:,i])figure(figsize=(22,4))    plt.subplot(1,3,1)plt.plot(r, np.abs(inverse), 'black', label = 'reconstruida')plt.plot(r, data, 'green', label = 'real')plt.ylabel(u'Amplitud pulsos')plt.xlabel('Tiempo (minutos)')plt.subplot(1,3,2)plt.plot(r, np.abs(inverse[:,0]), 'black', label = 'reconstruida')plt.plot(r, data[:,0], 'green', label = 'real')plt.ylabel(u'Amplitud pulsos')plt.xlabel('Tiempo (minutos)')savefig('inversa1.jpeg')plt.subplot(1,3,3)plt.plot(r, np.abs(inverse[:,20]), color = 'black', label = 'reconstruida')plt.plot(r,data[:,20], 'green', label = 'real')plt.ylabel(u'Amplitud pulsos')plt.xlabel('Tiempo (minutos)')savefig('inversa21.jpeg')
```

Ahora hallaremos el valor de chi² para cada señal y así poder tener una comparación cuantitativa entre las señales originales y las señales reconstruídas

```python
temp = zeros(col)chi2 = empty(col)for i in range(col):                            #loop para hallar el chi² de cada señal    for j in range(n):        temp[i] += (data[j,i]-np.abs(inverse[j,i]))**2        for i in range(col):    chi2[i] = temp[i]/nfor i in range(col):    print "chi2[" + str(i) + "] = ", chi2[i]    #Los valores de chi² de cada señal se imprimen                                                #de la siguiente manera: chi2[0] corresponde al                                                 #chi2 de la señal 1, chi2[1], corresponde al chi2                                                 #de la señal 2, etc.
```
