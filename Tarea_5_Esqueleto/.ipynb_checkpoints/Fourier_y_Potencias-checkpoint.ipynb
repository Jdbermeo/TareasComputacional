```python
import numpy as np
from scipy.fftpack import fft, fftfreq,ifft
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

data = np.loadtxt("datosTransConstante.txt")
mediciones = data[:,1]
dt = 1/12.0
fft_x = fft(mediciones)
fft_x = fft_x/len(mediciones)
freq = fftfreq(len(mediciones), dt)
plt.plot(freq,np.abs(fft_x)**2)
```

```python
#Filtra las potencias que cuyo Periodo sea mayor a 20 años o menor a 2 años
for i in range(0,len(freq)):
    if(freq[i]**2>0.5**2):
            fft_x[i] = 0
           

for i in range(0,len(freq)):
    if(freq[i]**2<0.05**2):
            fft_x[i] = 0
```

```python
potencia_espec = np.abs(fft_x)**2
plot(freq,np.abs(fft_x)**2)
mayor = 0 
ind_mayor = 0
for i in range(0,len(potencia_espec)):
    if(potencia_espec[i]>mayor):
        mayor = potencia_espec[i]
        ind_mayor = i
        
print "El periodo del ciclo solar es de ", 1/freq[ind_mayor], "años"
```

```python
#Manchas reconstruidas (debe multiplicarse por el numero de puntos otra vez?)
manchas_recons = ifft(fft_x*len(mediciones))
```

```python
scatter(data[:,0],mediciones)
plot(data[:,0],np.abs(manchas_recons),color='green')
legend(['Reconstruccion Constante','true'])
plt.ylim(-100,600)
```

```python
plt.title("Error porcentual")
plot(data[:,0],(((np.abs(manchas_recons)-mediciones)/mediciones)*100))
```
