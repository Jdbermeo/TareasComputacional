Primera Aproximacion

```python
n_x = 200
c = 1.0
dx = 2.0/n_x
beta = 0.01
dt = beta*dx 

n_t = int(0.15/dt) 
x = linspace(0, 2.0, n_x)
u = 3.0*sin(pi*x)
t = linspace(0, 0.15, 200)
plot(x,u)
X = numpy.array(x)
T = numpy.array(t)
```

```python
U = u.reshape(n_x,1)
for n in range(n_t):  # loop over time
    u_past = u.copy() 
    for i in range(1,n_x-1): #loop over space
        u[i] = u_past[i]  - beta*((u_past[i+1])**2-(u_past[i-1])**2)*0.5
    if(i<=199):  
        U = np.vstack([U, u.reshape(n_x,1)])

plot(x,u)    
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
for i in range(len(U)):
    ax.plot_wireframe(X,T,U[i])
```

Esto se debe a la estructura matemática del método que toma valores anteriores y posteriores y los promedia, siendo evidente que ante cambios bruscos de pendiente, se verá afectado el valor de la función por el mismo hecho de que está variando muy rápidamente. Por su parte, el método de Lax–Wendroff, si bien también se basa en el método de diferencias finitas, hace una mejor aproximación al tomar intervalos más pequeños y por ende promedios más cercanos que le dan un grado de detalle mucho mayor a la solución numérica. En este caso el proceso es más largo pero se asegura una buena aproximación.
